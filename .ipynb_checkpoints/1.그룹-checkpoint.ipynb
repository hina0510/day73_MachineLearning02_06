* agg : 집계 연산

```python
import numpy as np
df_class.agg({'math':np.max, 'english':np.min})
```

```python
import numpy as np
df_class.agg({'math':np.mean, 'english':np.sum})
```

```python
import numpy as np
df_class.agg({'math':[np.mean, np.sum], 'english':np.sum})
```
